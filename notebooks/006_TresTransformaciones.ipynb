```python
import numpy as np
import matplotlib.pyplot as plt
%run "./funcionesAuxiliares/graficarVectores.ipynb"
%run "./funcionesAuxiliares/graficarMatriz.ipynb"
```

```python
A = np.array([[3,7], [5,2]])
print(A)
```

```python
print("Circulo unitario")
graficarMatriz(np.array([[1,0], [0,1]]))


plt.xlim(-1.5, 1.5)
plt.ylim(-1.5, 1.5)
```

```python
print("Circulo unitario transformado")
graficarMatriz(A)


plt.xlim(-8, 8)
plt.ylim(-8, 8)
```

```python
U, D, V = np.linalg.svd(A)
print(U)
```

```python
print("Circulo unitario matriz U")
graficarMatriz(U)

plt.xlim(-1.5, 1.5)
plt.ylim(-1.5, 1.5)


print("Primer rotacion V: ")
graficarMatriz(V)

plt.xlim(-1.5, 1.5)
plt.ylim(-1.5, 1.5)

print("Escala (D):")
graficarMatriz(np.diag(D) @ V)

plt.xlim(-9, 9)
plt.ylim(-9, 9)

print("Segunda rotacion U: ")
graficarMatriz(U.dot(np.diag(D).dot(V)))

plt.xlim(-8, 8)
plt.ylim(-8, 8)
```
