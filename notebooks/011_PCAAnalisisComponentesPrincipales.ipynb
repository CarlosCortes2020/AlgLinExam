```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
np.random.seed(42)

x = 3*np.random.rand(200)
y =20*x + 2*np.random.randn(200)

x = x.reshape(200,1)
y = y.reshape(200,1)
print(x)
print(y)
```

```python
xy = np.hstack([x,y])
print(xy.shape)
```

```python
plt.plot(xy[:,0], xy[:,1], "." )
plt.show()
```

```python
xy_cent = xy - np.mean(xy, axis = 0)
plt.plot(xy_cent[:,0], xy_cent[:,1], "." )
plt.show()
```

```python
autovalores, autovectores = np.linalg.eig(xy_cent.T.dot(xy_cent))
print("Autovalores: \n", autovalores)
print("Autovectores: \n", autovectores)
```
