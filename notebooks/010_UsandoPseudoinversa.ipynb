```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.linspace(-5, 5, 1000)

y1 = -4*x + 3 
y2 = 2*x + 5
y3 = -3*x + 1

plt.plot(x, y1, label='y1 = -4x + 3')
plt.plot(x, y2, label='y2 = 2x + 5')
plt.plot(x, y3, label='y3 = -3x + 1')

plt.xlim(-5, 5)
plt.ylim(-5, 5)
```

```python
#Determinar el punto que minimiza la norma 2
matriz = np.array([[4, 1], [-2, 1], [3, 1]])
print("Matriz de coeficientes:\n", matriz)
```

```python
matriz_pse = np.linalg.pinv(matriz)
print("Matriz pseudoinversa: \n", matriz_pse)
```

```python
b = np.array([[3], [5], [1]])
print("Vector de terminos independientes: \n", b)
```

```python
resultado = matriz_pse.dot(b)
print("El punto que minimiza la norma 2 es: \n", resultado)
```
